```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
df = pd.read_csv("student_scores_data.csv")
df
```

```python
plt.figure (figsize=(12,8))
plt.scatter(x=df["SAT"],
           y=df["GPA"],
           color = "darkblue",
           alpha= 0.6)

plt.title("Relationship between SAT & GPA scores of student",
          fontsize = 16,
          weight = "bold",
          color = "darkblue")
plt.xlabel("SAT Scores", 
           fontsize = 13,
           weight = "bold",
           color = "darkblue")
plt.ylabel("GPA Scores",
           fontsize = 13,
           weight ="bold",
           color = "darkblue")
plt.show()
```

```python
plt.figure (figsize=(12,8))
plt.scatter(x=df["SAT"],
           y=df["GPA"],
           color = "darkblue",
           alpha= 0.6)

plt.title("Relationship between SAT & GPA scores of student",
          fontsize = 16,
          weight = "bold",
          color = "darkblue")
plt.xlabel("SAT Scores", 
           fontsize = 13,
           weight = "bold",
           color = "darkblue")
plt.ylabel("GPA Scores",
           fontsize = 13,
           weight ="bold",
           color = "darkblue")
plt.xlim(0, max(df["SAT"]+500))
plt.ylim(0, max(df["GPA"]+1))
plt.show()
```
