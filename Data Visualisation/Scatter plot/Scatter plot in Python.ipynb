```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
df = pd.read_csv("scatter_data.csv")
df
```

```python
plt.figure(figsize=(12,8))
#sns.set_style("white")
scatter  = plt.scatter(df["Area (ft.)"],
           df["Price"],
           alpha=0.5,
           c = df["Building Type"],
           cmap="viridis")
plt.legend(*scatter.legend_elements(), 
          loc = "upper left",
          title = "Building Type")

plt.title("Relationship between Area and Prices of California Real Estate",
         fontsize = 16,
         weight ="bold",
         color = "navy",
         loc = "left")
plt.xlabel("Areas (sq. ft)", 
          fontsize = 12,
          weight = "bold",
          color = "navy")
plt.ylabel("Prices (K$)",
          fontsize = 12,
          weight = "bold",
          color = "navy")
plt.xticks(color = "navy", 
           weight = "bold")
plt.yticks(color = "navy",
          weight = "bold")
sns.despine()
plt.show()
```

```python
plt.figure(figsize=(12,8))
#sns.set_style("white")

sns.scatterplot(x=df["Area (ft.)"],
               y=df["Price"],
               hue=df["Building Type"],
               palette=['black', 'darkblue', 'purple', 'pink', 'white'],
               s = 100)

plt.title("Relationship between Area and Prices of California Real Estate",
         fontsize = 16,
         weight ="bold",
         color = "navy",
         loc = "left")
plt.xlabel("Areas (sq. ft)", 
          fontsize = 12,
          weight = "bold",
          color = "navy")
plt.ylabel("Prices (K$)",
          fontsize = 12,
          weight = "bold",
          color = "navy")
plt.xticks(color = "navy", 
           weight = "bold")
plt.yticks(color = "navy",
          weight = "bold")
sns.despine()
plt.show()
```
