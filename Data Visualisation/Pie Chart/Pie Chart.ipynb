```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
df = pd.read_csv("pie_chart_data.csv")
df
```

```python
plt.figure(figsize=(10,6))
plt.pie(df["Number of Cars"],
        labels=df["Engine Fuel Type"], 
        autopct="%1.1f%%",
        textprops ={'color': "w", 'fontweight' :"bold"})
plt.legend()
plt.title('Cars by Fuel Engine Type', fontsize = 16, fontweight= "bold")
plt.show()
```
