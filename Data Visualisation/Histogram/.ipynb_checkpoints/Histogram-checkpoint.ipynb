```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("histogram_data.csv")
df
```

```python
sns.set_style("white")

labels = ("[108,162]", "(162,216]", "(216,270]", "(270,324]","(324,378]","(378,432]", "(432,486]","(486,540]")

plt.figure(figsize=(12,8))
plt.hist(df["Price"], bins=8)
plt.title("Distribution of Real Estate Prices", fontsize = 14, fontweight = "bold", color = "darkblue", loc = "left")
plt.ylabel("Nubmer of house", fontsize = 12, fontweight="bold", color = "darkblue")
plt.xlabel("Prices in K$", fontsize = 12, fontweight = "bold", color = "darkblue")
plt.xlim(108, 540)
sns.despine()
plt.show()
```
