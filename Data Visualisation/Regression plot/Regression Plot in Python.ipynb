```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df  = pd.read_csv("scatter_plot_ii.csv")
df
```

```python
plt.figure(figsize=(10,8))
sns.regplot(data=df,
           x="Budget",
           y="Sales",
           scatter_kws = {"color": "black"},
           line_kws = {"color": "red"})
plt.xlabel("Ad Expenditure (K$)",
          fontsize = 12)
plt.ylabel("Sales (KUnits)",
          fontsize = 12)
plt.title("Effect of Ad Expenditure on Sales", 
         fontsize = 14,
         weight = "bold")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.lmplot(data=df,
           x="Budget",
           y="Sales",
           scatter_kws = {"color": "black"},
           line_kws = {"color": "red"},
          height=9)
plt.xlabel("Ad Expenditure (K$)",
          fontsize = 12)
plt.ylabel("Sales (KUnits)",
          fontsize = 12)
plt.title("Effect of Ad Expenditure on Sales", 
         fontsize = 14,
         weight = "bold")
plt.show()
```
