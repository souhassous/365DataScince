```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("area_chart_homework_data.csv")
df
```

```python
df["Date"] = pd.to_datetime(df["Date"], format="%m/%d/%Y")
df
```

```python
colors = ["#011638", "#7e2987", "#ef2026"]
labels = ["International", "Services", "Domestic"]
sns.set_style("white")
plt.figure(figsize=(12, 6))
plt.stackplot(df["Date"],
              df["International"],
              df["Services"],
              df["Domestic"],
              colors=colors,
              edgecolor="none")
plt.xticks(df["Date"], rotation=45)
plt.legend(labels=labels, loc="upper left")
plt.title("Revenue Report 2020", fontsize=18,
          weight="bold", loc="left", color=colors[0])
plt.xlabel("Date", fontsize=13)
plt.ylabel("Revenue in $", fontsize=13)
sns.despine()
plt.show()
```
