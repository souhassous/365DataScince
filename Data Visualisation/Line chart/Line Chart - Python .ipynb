```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
df = pd.read_csv("line_chart_data.csv")
df
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df
```

```python
startDate = "2008-07-01"
endDate = "2008-12-31"
df = df[(df.Date >= startDate) & (df.Date <= endDate)]
df
```

```python
labels = ["S&P 500", "FTSE 100"]
plt.figure(figsize=(20,8))
plt.plot(df["Date"], df["GSPC500"], color = "midnightblue")
plt.plot(df["Date"], df["FTSE100"], color = "crimson")
plt.title("S&P vs FTSE (H2 2008)", fontsize = 16, loc="left", color = "darkblue", fontweight = "bold")
plt.xlabel("Date", fontsize = 14, color = "darkblue", fontweight = "bold")
plt.ylabel("Returns", fontsize = 14, color = "darkblue", fontweight = "bold")
plt.legend(labels = labels, fontsize ="large")
plt.show()
```
